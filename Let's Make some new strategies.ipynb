```python
import pandas as pd
import numpy as np
import mlxtend 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
```

```python

with open("C:/Users/Aya Zaoui/Downloads/Market_Basket_Optimisation.csv",'r')as f:
    dataset=[[w.strip() for w in line .strip().split(',')]for line in f.readlines()]
```

```python
te=TransactionEncoder()
df=pd.DataFrame(te.fit_transform(dataset),columns=te.columns_).astype(int)
df.head()
```

```python
plt.figure(figsize=(18,12))
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(18,12))
df.sum().sort_values(ascending=False).plot.bar()
```

```python
from mlxtend.frequent_patterns import * 
apr=apriori(df, min_support=0.05).sort_values('support', ascending=False)
association_rules(apr,metric='confidence', min_threshold=0.1)
```
